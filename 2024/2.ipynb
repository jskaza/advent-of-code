```python
import numpy as np
```

```python
inputs = open('data/2.txt').read()
```

Part 1

```python
def safety_check(arr):
  return np.all((np.diff(arr) >= 1) & (np.diff(arr) <= 3)) or np.all((np.diff(arr) >= -3) & (np.diff(arr) <= -1))
sum([safety_check(np.array(x.split()).astype(int)) for x in inputs.splitlines()])
```

Part 2

```python
def safety_check2(arr):
  if np.all((np.diff(arr) >= 1) & (np.diff(arr) <= 3)) or np.all((np.diff(arr) >= -3) & (np.diff(arr) <= -1)):
    return True
  else:
    for i in range(len(arr)):
      temp = np.delete(arr, i)
      if safety_check(temp):
        return True
  return False
sum([safety_check2(np.array(x.split()).astype(int)) for x in inputs.splitlines()])
```
