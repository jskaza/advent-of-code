```python
data = [x.strip().replace(",", "|").split("|") for x in open('data/5.txt').readlines()]
```

Part 1

```python
rules = {}
updates = []
rules_flag = True
for line in data:
    if line[0] == "":
        rules_flag = False
    elif rules_flag:
        if line[0] in rules:
            rules[line[0]].append(line[1])
        else:
            rules[line[0]] = [line[1]]
    else:
        updates.append(line)

def check_update(update, rules):
    for i,u in enumerate(update):
      if u not in rules:
          continue
      else:
          if set(update[:i]) & set(rules[u]):
            return 0
    return int(update[i//2])

sum([check_update(update, rules) for update in updates])
```

Part 2

```python
def check_update(update, rules):
    for i,u in enumerate(update):
      if u not in rules:
          continue
      else:
          if set(update[:i]) & set(rules[u]):
            return 0
    return int(update[i//2])

def order_update(update, rules):
    i = 0
    while i < len(updates):
      if update[i] not in rules:
       i+=1
      else:
        for j in range(len(update[:i])):
            if update[j] in rules[update[i]]:
                update.insert(i+1, update[j])
                del update[j]
        i = 0
    return int(update[i//2])

sum([order_update(update, rules) for update in updates if not check_update(update, rules)])
```
