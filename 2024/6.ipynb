```python
with open('data/6.txt', 'r') as file:
    lines = file.readlines()
grid = [x.strip() for x in lines]
```

Part 1

```python
visited = set()
for r, row in enumerate(grid):
    for c, col in enumerate(row):
        if col in ["^",">","<","v"]:
            pos = {"row": r, "col": c}
            state = col
            break
visited.add(tuple(pos.values()))
grid[pos["row"]] = grid[pos["row"]].replace(state, ".")
while True:          
    if state == "^" and grid[pos["row"]-1][pos["col"]] == ".":
        pos["row"]-=1
    elif state == "^" and grid[pos["row"]-1][pos["col"]] == "#":
        state = ">"
        pos["col"]+=1
    elif state == "v" and grid[pos["row"]+1][pos["col"]] == ".":
        pos["row"]+=1
    elif state == "v" and grid[pos["row"]+1][pos["col"]] == "#":
        state = "<"
        pos["col"]-=1
    elif state == ">" and grid[pos["row"]][pos["col"]+1] == ".":
        pos["col"]+=1
    elif state == ">" and grid[pos["row"]][pos["col"]+1] == "#":
        state = "v"
        pos["row"]+=1
    elif state == "<" and grid[pos["row"]][pos["col"]-1] == ".":
        pos["col"]-=1
    elif state == "<" and grid[pos["row"]][pos["col"]-1] == "#":
        state = "^"
        pos["row"]-=1
    if pos["row"] < 0 or pos["row"] >= len(grid) or pos["col"] < 0 or pos["col"] >= len(grid[0]):
        break
    visited.add(tuple(pos.values()))
print(len(visited))
```
