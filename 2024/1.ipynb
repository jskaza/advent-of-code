```python
import numpy as np
from collections import Counter
```

```python
data = open('data/1.txt').read()
```

Part 1

```python
np.sum(np.abs(np.array(sorted(map(int, data.split()[::2]))) - np.array(sorted(map(int, data.split()[1::2])))))
```

Part 2

```python
left_list = map(int, data.split()[::2])
right_list = Counter(map(int, data.split()[1::2]))
def f(x):
  if x in right_list:
    return x*right_list[x]
  else:
    return 0
np.sum(list(map(f, left_list)))
```
