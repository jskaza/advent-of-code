```python
import re
import math
```

```python
inputs = open('data/3.txt').read()
```

Part 1

```python
sum([math.prod(map(int, re.findall(r"\d+", x))) for x in re.findall(r"mul\(\d+,\d+\)", inputs)])
```

Part 2

```python
inputs2 = "do()" + inputs
muls = {int(match.start()): match.group() for match in re.finditer(r"mul\(\d+,\d+\)", inputs2)}
conditions = {int(match.start()): match.group() for match in re.finditer(r"do\(\)|don't\(\)", inputs2)}

def find_closest_prior_index(target_index, conditions):
    prior_indices = [i for i in conditions if i < target_index]
    return conditions[max(prior_indices)]

sum([math.prod(map(int, re.findall(r"\d+", v))) for k,v in muls.items() if find_closest_prior_index(k,conditions) == "do()"])
```
