```python
import numpy as np
import re
```

```python
data = np.array([list(x.strip()) for x in open('data/4.txt').readlines()])
```

Part 1

```python
rows = [''.join(data[i, :]) for i in range(data.shape[0])]
cols = [''.join(data[:, j]) for j in range(data.shape[1])]
main_diagonal = [''.join(np.diag(data))]
upper_diagonals = [''.join(np.diag(data, k)) for k in range(1, data.shape[1])]
lower_diagonals = [''.join(np.diag(data, -k)) for k in range(1, data.shape[0])]
flipped_array = np.fliplr(data)
anti_diagonals = [''.join(np.diag(flipped_array, k)) for k in range(-data.shape[0] + 1, data.shape[1])]
all_lines = rows + cols + main_diagonal + upper_diagonals + lower_diagonals + anti_diagonals

sum([len(re.findall(r"XMAS", line)) + len(re.findall(r"SAMX", line)) for line in all_lines])
```

Part 2

```python
xmas = 0
for r in range(len(data)):
    for c in range(len(data[0])):
        current_char = data[r][c]
        if current_char != "A":
            continue
        else:
            try:
                diagonal1 = f"{data[r-1][c-1]}{current_char}{data[r+1][c+1]}"
                diagonal2 = f"{data[r+1][c-1]}{current_char}{data[r-1][c+1]}"
                if diagonal1 in ["MAS", "SAM"] and diagonal2 in ["MAS", "SAM"]:
                    xmas += 1
            except:
                continue
xmas
    
```
